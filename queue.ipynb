```python
class fifo():
    
    def __init__(self):
        
        self.items = []
        
    def push(self, item):
        self.items.insert(0, item)
        
    def pop(self):
        name = self.items[-1]
        self.items.remove(name)
        return name
        
        
    def peek_last(self):
        return self.items[0]
    
    def peek_first(self):
        return self.items[-1]
        
    def show(self):
        print(self.items)
        
    def is_empty(self):
        if (len(self.items) == 0):
            return True
        else:
            return False
        
    def length(self):
        return len(self.items)
        
def hot_pot(names, n):
    
    fqu = fifo()
    for name in names:
        fqu.push(name)
        
    while (fqu.length() > 1):
        for i in range(n):
            head_name = fqu.pop()
            fqu.push(head_name)
            #fqu.show()
            
        fqu.pop()
    
    return fqu.peek()
```

```python
name = hot_pot(names, 5)
print name
```

```python
import numpy as np
from time import time
```

```python
def simulate(time_duration):
    
    tick = time()
    old_print_que = fifo()
    new_print_que = fifo()
    old_print_wait_sum = 0
    new_print_wait_sum = 0
    old_print_que_count = 0
    new_print_que_count = 0
    
    while (time() <= (tick + time_duration)):
        
        #check both printer ques to see if a job is done
        #if it is, then pop it from the que
        if (not old_print_que.is_empty()):
            first_item = old_print_que.peek_first()
            if (time() > first_item[0]+first_item[1]):
                old_print_wait_sum += time()-first_item[0]
                old_print_que.pop()
        
        if (not new_print_que.is_empty()):
            first_item = new_print_que.peek_first()
            if (time() > first_item[0] + first_item[1]):
                new_print_wait_sum += time()-first_item[0]
        
        #randomly create a print job to put on both printer ques
        event = np.random.randint(low = 1, high = 181)
        if (event == 180):
            #print ("Create random for old printer")
            numb_of_pages = np.random.randint(low = 1, high = 21)
            old_printer_job_time = float(numb_of_pages)/6
            old_print_que.push([time(), old_printer_job_time])
            old_print_que_count += 1
            #print ("old:", numb_of_pages, old_printer_job_time, old_print_que_count)
            #old_print_que.show()
            #print("")
        
        event = np.random.randint(low = 1, high = 181)
        if (event == 180):
            #print ("Create random for new printer")
            numb_of_pages = np.random.randint(low = 1, high = 21)
            new_printer_job_time = float(numb_of_pages)/12
            new_print_que.push([time(), new_printer_job_time])
            new_print_que_count += 1
            #print ("new:", numb_of_pages, new_printer_job_time, new_print_que_count)
            #new_print_que.show()
            #print("")
            
    #empty both the que's
    while((not old_print_que.is_empty()) or (not new_print_que.is_empty())):
        if (not old_print_que.is_empty()):
            first_item = old_print_que.peek_first()
            if (time() > first_item[0] + first_item[1]):
                old_print_wait_sum += time() - first_item[0]
                old_print_que.pop()
                
        if (not new_print_que.is_empty()):
            first_item = new_print_que.peek_first()
            if (time() > first_item[0] + first_item[1]):
                new_print_wait_sum += time() - first_item[0]
                new_print_que.pop()

    print ("The average wait time on the old printer queue is " + str(old_print_wait_sum/old_print_que_count) + " seconds.")
    print ("The average wait time on the new printer queue is " + str(new_print_wait_sum/new_print_que_count) + " seconds.")
```

```python
for i in range(5):
    simulate(5)
```
